# DEMOGRAPHIC ANALYSIS

## Age distribution of consumers.

```python
plt.figure(figsize = (8,5))
sns.histplot(data['Age'],bins=20, kde=True)
plt.title('Age distribution of consumers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

* The age group distribution seems to be relatively uniform across the dataset.
* There are no significant peaks of dominating any specific age group.
* There are slight variations with minor peaks around 20 - 30 and 50 - 60 age group, suggesting a slightly higher number of conumers is these age group.

## Shopping behaviors between male and female consumers.

```python
print(data['Gender'].value_counts())

# Analyze average purchase amount by gender
avg_purchase_amount = data.groupby('Gender')['Purchase Amount (USD)'].mean()
print("Average Purchase Amount by Gender:")
print(avg_purchase_amount)

# Analyze frequency of purchases by gender
purchase_frequency = data.groupby('Gender')['Customer ID'].count()
print("Purchase Frequency by Gender:")
print(purchase_frequency)

# Analyze preferred product categories by gender
category_preference = data.groupby(['Gender', 'Category'])['Customer ID'].count().unstack().fillna(0)
print("Product Category Preference by Gender:")
print(category_preference)
```

* The average amount spent on purchase are very close to each other, suggesting both male and female consumers tend to spend
similar amount.
* The purchase frequency of male is more than the female consumers, since the average purchase amount is similar, increasing
the purchase frequency amount among female consumers can boost the overall revenue.
* The analysis reveals that males tend to purchase more across all product categories compared to females, with clothing 
being the most popular category for both genders.

# PRODUCT AND CATEGORY INSIGHTS

## Top selling product in each category.

```python
top_selling_product = data.groupby(['Category', 'Item Purchased'])['Purchase Amount (USD)'].sum().reset_index()
top_selling_product = top_selling_product.loc[top_selling_product.groupby('Category')['Purchase Amount (USD)'].idxmax()]
print('The top selling product in each category:')
print(top_selling_product)
```

```python
plt.figure(figsize=(8, 4))
plt.bar(top_selling_product['Category'], top_selling_product['Purchase Amount (USD)'], color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Category')
plt.ylabel('Purchase Amount')
plt.title('Top Selling in Each Category')
plt.show()
```
